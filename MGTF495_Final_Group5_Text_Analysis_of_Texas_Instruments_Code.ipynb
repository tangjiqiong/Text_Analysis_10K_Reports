#### For text analysis, we use the ‘fitz’ module to extract text by pages from the PDF document and expand each page’s text into one list named ‘text’.

```python
import fitz

filename = 'TI20.pdf'
doc = fitz.open(filename)

pages = 0
text = []
for i in range(doc.page_count):
    pages = doc.load_page(i)
    text += [pages.get_text('text')]

print(text)
```

#### To better get important phrases and sentences, define a search_by_regexp(regexp)to output the target page and text contents by searching for keywords using the ‘re’ module.

```python
#search function
import re

def search_by_regexp(regexp):
    for i in range(doc.page_count):    
        raw_doc = text[i]
        mytokens = tk.sent_tokenize(raw_doc) 
        for l in mytokens:
            x = re.search(regexp, l)
            if x and not '10-K' in l:
                print(f'page{i}')
                print(mytokens.index(l), ': ', l)
```

```python
def search_by_regexp_ori(regexp, section=None):
    if section:
        raw_doc = section
        mytokens = tk.sent_tokenize(raw_doc) 
        trigger = False
        for l in mytokens:
            x = re.search(regexp, l)
            if x and not '10-K' in l:
                trigger = True
            if trigger:
                print(l)
    else:
        for i in range(doc.page_count):    
            raw_doc = text[i]
            mytokens = tk.sent_tokenize(raw_doc) 
            for l in mytokens:
                x = re.search(regexp, l)
                if x and not '10-K' in l:
                    print(f'page{i}')
                    print(mytokens.index(l), ': ', l)
```

### Question 1
#### What is the total number of shares that TI repurchased on its own common stocks and what is the average cost of per share, comparing 5 years' 10K report.

#### The steps to reach the results

1.Search for  'repurchase \d' in every 10-k to check if there is specific text pattern about stock repurchasing.

2.Use re.compile to search for my keywords and use ‘findall’ to get all my results.

3.Extract the numbers of the total cost and shares and calculate the average cost per share.

4.Loop for five years:

Filenames for these five reports are saved as 'TI 2020.pdf', 'TI 2019.pdf' and so on.


```python
search_by_regexp('repurchase \d')
```

```python
shares_list = []
dollar_list = []
avg_cost_list = []
for l in range(16,21):
    
    filename = 'TI' + str(l) + '.pdf'
    doc = fitz.open(filename)

    pages = 0
    text = []
    for i in range(doc.page_count):
        pages = doc.load_page(i)
        text += [pages.get_text('text')]

    find_shares = re.compile('repurchase (\d*[.]\d*)')
    shares = find_shares.findall(str(text))[0]
    shares_list.append(float(shares))

    find_dollar = re.compile('(\d*[.]\d*) billion.to repurchase')
    dollar = find_dollar.findall(str(text))[0]
    dollar_list.append(float(dollar))
    avg_cost = float(dollar)*1000 / float(shares)
    avg_cost_list.append(round(avg_cost,2))
print('The total cost (in billion) of stock repurchasing for TI in every year : ''\n', dollar_list)
print('The total shares (in million) that TI repurchased in every year :''\n', shares_list)
print('The average price per share(in $) of stock repurchasing for TI in every year :''\n', avg_cost_list)
```

### Question5 
#### How much has TI's R&D spending been in the past three years? How much is the proportion of R&D expenditures to Texas Instrument’s profits?

#### The steps to reach the results

1.Get the target page:

Using the re_search function, find the pages which have the table ‘Consolidated Statements of Income’, we can get that the table is on page 25.

2.Extract the table and transfer to dataframe:

Using the camelot module to extract the target table. Then using pandas package to polish the dataframe. After deleting the first two rows, we set the first column as row indexes and the header as year 2020, 2019, and 2018.

3.Calculating the proportion of  R&D Expenditure in Net income:

Since the type of numbers we extracted from the original pdf document is string with comma, we are not able to calculate the proportion directly. We make arrays and for-loops to do the calculation then append the new array ‘proportion’ as the third row into our result.


```python
search_by_regexp('Consolidated Statements of Income')  
```

```python
import camelot
from camelot import read_pdf
tables = camelot.read_pdf(filename, pages = '25', flavor = 'stream')
a = tables[0].df
Income_St = a.drop(index = [0,1], columns = [1,3,5])
Income_St.columns = ['Accounting Subjects','2020','2019','2018']
Income_St.set_index(['Accounting Subjects'],inplace=True)
print(Income_St)
```

```python
result = Income_St.loc[['Research and development (R&D)','Net income']]
result = result[0:2]

RnD_all =[]
Net_income_all = []

for n in range(result.shape[1]): 
    RnD_temp = result.loc['Research and development (R&D)'][n]  #loc[row],[column]
    RnD = float(RnD_temp.replace(",",""))  ## delete the comma ","
    RnD_all.append(RnD)
    
    Net_income_temp = result.loc['Net income'][n]
    Net_income = float(Net_income_temp.replace(",",""))  ## delete the comma ","
    Net_income_all.append(Net_income)
RnD_all = np.array(RnD_all)
Net_income_all = np.array(Net_income_all)

propotion = RnD_all / Net_income_all
result.loc['Proportion'] = propotion
print(result)
```
